```python
import os
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.model_selection import StratifiedKFold
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import classification_report, confusion_matrix, ConfusionMatrixDisplay, roc_curve, auc
import umap
import torch
import torch.nn as nn
import torch.optim as optim
from torch.utils.data import Dataset, DataLoader

# Create directories for saving outputs
BASE_DIR = "/content/drive/MyDrive/ddos_results"
MODEL_DIR = os.path.join(BASE_DIR, "models")
PLOTS_DIR = os.path.join(BASE_DIR, "plots")
REPORTS_DIR = os.path.join(BASE_DIR, "reports")

os.makedirs(MODEL_DIR, exist_ok=True)
os.makedirs(PLOTS_DIR, exist_ok=True)
os.makedirs(REPORTS_DIR, exist_ok=True)

# Load dataset
df = pd.read_csv("/content/drive/MyDrive/shuffled_merged_data.csv", low_memory=False)
df['Label'] = df['Label'].replace({'1': 1, 'Benign': 0}).astype(int)

selected_features = [
    'Protocol', 'Flow Duration', 'Fwd Packet Length Max', 'Bwd Packet Length Max',
    'Flow Bytes/s', 'Flow Packets/s', 'Packet Length Mean', 'ACK Flag Count',
    'Idle Mean', 'Idle Std', 'Fwd IAT Max', 'Bwd IAT Max', 'Fwd PSH Flags',
    'Fwd URG Flags', 'Fwd Header Length', 'Packet Length Variance'
]

X = df[selected_features].fillna(0).astype(np.float32)
y = df['Label'].values

correlations = df[selected_features + ['Label']].corr()['Label'].abs().sort_values(ascending=False)
print("\nTop correlated features with label:\n", correlations)

scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

class AckDataset(Dataset):
    def __init__(self, X, y):
        self.X = torch.tensor(X, dtype=torch.float32)
        self.y = torch.tensor(y, dtype=torch.float32)
    def __len__(self): return len(self.X)
    def __getitem__(self, idx): return self.X[idx], self.y[idx]

class ACKTransformer(nn.Module):
    def __init__(self, input_dim):
        super().__init__()
        self.input_proj = nn.Linear(input_dim, 64)
        encoder_layer = nn.TransformerEncoderLayer(d_model=64, nhead=4, dim_feedforward=128, batch_first=True)
        self.transformer = nn.TransformerEncoder(encoder_layer, num_layers=2)
        self.fc = nn.Sequential(
            nn.Linear(64, 32), nn.ReLU(), nn.Dropout(0.2), nn.Linear(32, 1), nn.Sigmoid()
        )
    def forward(self, x):
        x = self.input_proj(x).unsqueeze(1)
        x = self.transformer(x).squeeze(1)
        return self.fc(x)

skf = StratifiedKFold(n_splits=5, shuffle=True, random_state=42)
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")

fold = 1
for train_idx, val_idx in skf.split(X_scaled, y):
    print(f"\n游대 Fold {fold}/5")

    X_train, X_val = X_scaled[train_idx], X_scaled[val_idx]
    y_train, y_val = y[train_idx], y[val_idx]

    train_loader = DataLoader(AckDataset(X_train, y_train), batch_size=512, shuffle=True)
    val_loader = DataLoader(AckDataset(X_val, y_val), batch_size=512)

    model = ACKTransformer(input_dim=X.shape[1]).to(device)
    criterion = nn.BCELoss()
    optimizer = optim.Adam(model.parameters(), lr=1e-3)

    best_f1, patience, best_state = 0, 0, None
    train_losses, val_losses = [], []
    train_accuracies, val_accuracies = [], []
    val_f1s = []

    for epoch in range(1, 51):
        model.train()
        epoch_loss = 0
        train_preds, train_labels = [], []

        for xb, yb in train_loader:
            xb, yb = xb.to(device), yb.to(device).unsqueeze(1)
            optimizer.zero_grad()
            out = model(xb)
            loss = criterion(out, yb)
            loss.backward()
            optimizer.step()

            epoch_loss += loss.item()
            preds = (out > 0.5).int()
            train_preds.extend(preds.cpu().numpy())
            train_labels.extend(yb.cpu().numpy())

        train_report = classification_report(train_labels, train_preds, output_dict=True, zero_division=0)

        model.eval()
        val_preds, val_labels, val_probs = [], [], []
        val_loss = 0
        with torch.no_grad():
            for xb, yb in val_loader:
                xb, yb = xb.to(device), yb.to(device).unsqueeze(1)
                out = model(xb)
                loss = criterion(out, yb)
                val_loss += loss.item()
                preds = (out > 0.5).int()
                val_preds.extend(preds.cpu().numpy())
                val_labels.extend(yb.cpu().numpy())
                val_probs.extend(out.cpu().numpy())

        val_report = classification_report(val_labels, val_preds, output_dict=True, zero_division=0)
        fpr, tpr, _ = roc_curve(val_labels, val_probs)
        roc_auc = auc(fpr, tpr)

        train_losses.append(epoch_loss)
        val_losses.append(val_loss)
        train_accuracies.append(train_report['accuracy'])
        val_accuracies.append(val_report['accuracy'])
        val_f1s.append(val_report['weighted avg']['f1-score'])

        print(f"Epoch {epoch:02d}: Train Acc={train_report['accuracy']:.3f}, Val Acc={val_report['accuracy']:.3f} | "
              f"Prec={val_report['weighted avg']['precision']:.3f}, Rec={val_report['weighted avg']['recall']:.3f}, "
              f"F1={val_report['weighted avg']['f1-score']:.3f}, AUC={roc_auc:.3f}")

        if val_report['weighted avg']['f1-score'] > best_f1:
            best_f1 = val_report['weighted avg']['f1-score']
            best_state = model.state_dict()
            patience = 0
        else:
            patience += 1
            if patience >= 4:
                print(f"Early stopping at epoch {epoch}")
                break

    model.load_state_dict(best_state)
    torch.save(best_state, os.path.join(MODEL_DIR, f"best_model_fold_{fold}.pth"))

    # Save confusion matrix
    cm = confusion_matrix(val_labels, val_preds)
    ConfusionMatrixDisplay(confusion_matrix=cm).plot()
    plt.title(f"Confusion Matrix - Fold {fold}")
    plt.savefig(os.path.join(PLOTS_DIR, f"confusion_matrix_fold_{fold}.png"))

    # Training vs Validation metrics
    plt.figure(figsize=(12, 5))
    plt.subplot(1, 2, 1)
    plt.plot(train_losses, label='Train Loss')
    plt.plot(val_losses, label='Val Loss')
    plt.legend(); plt.title('Loss over Epochs')

    plt.subplot(1, 2, 2)
    plt.plot(train_accuracies, label='Train Acc')
    plt.plot(val_accuracies, label='Val Acc')
    plt.legend(); plt.title('Accuracy over Epochs')

    plt.savefig(os.path.join(PLOTS_DIR, f"train_val_metrics_fold_{fold}.png"))

    # ROC Curve
    plt.figure()
    plt.plot(fpr, tpr, label=f"AUC={roc_auc:.2f}")
    plt.plot([0, 1], [0, 1], 'k--')
    plt.xlabel('False Positive Rate')
    plt.ylabel('True Positive Rate')
    plt.title('ROC Curve')
    plt.legend()
    plt.savefig(os.path.join(PLOTS_DIR, f"roc_curve_fold_{fold}.png"))

    # UMAP Plot
    reducer = umap.UMAP(n_neighbors=15, min_dist=0.1, random_state=42)
    sample_idx = np.random.choice(len(X_val), size=min(5000, len(X_val)), replace=False)
    X_val_embedded = reducer.fit_transform(X_val[sample_idx])

    plt.figure()
    plt.scatter(X_val_embedded[:, 0], X_val_embedded[:, 1], c=y_val[sample_idx], cmap='coolwarm', alpha=0.5)
    plt.title(f"UMAP Visualization - Fold {fold}")
    plt.savefig(os.path.join(PLOTS_DIR, f"umap_plot_fold_{fold}.png"))

    report_df = pd.DataFrame(val_report).transpose()
    report_df.to_csv(os.path.join(REPORTS_DIR, f"report_fold_{fold}.csv"))

    fold += 1
```
